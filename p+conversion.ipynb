```python
import os
import pickle
import numpy as np
from PIL import Image

def create_pickle_files(image_folder, text_folder, output_folder):
    # Ensure the output folder exists
    os.makedirs(output_folder, exist_ok=True)

    # Loop through all files in the image folder
    for filename in os.listdir(image_folder):
        if filename.endswith(('.png', '.jpg', '.jpeg')):  # Add other extensions if needed
            # Load the image
            image_path = os.path.join(image_folder, filename)
            image = np.array(Image.open(image_path))

            # Extract the base filename without extension
            base_filename = os.path.splitext(filename)[0]

            # Read the corresponding text file
            text_file_path = os.path.join(text_folder, f'{base_filename}.txt')
            if os.path.exists(text_file_path):
                with open(text_file_path, 'r') as text_file:
                    text = text_file.read().strip()
            else:
                print(f'Text file not found for image {filename}. Skipping...')
                continue

            # Create the data dictionary
            data = {'img': image, 'text': text}

            # Save to .p file
            pickle_path = os.path.join(output_folder, f'{base_filename}.p')
            with open(pickle_path, 'wb') as file:
                pickle.dump(data, file)
            
            print(f'Created: {pickle_path}')

# Define the image folder, text folder, and output folder
image_folder = 'D:/Frank/8-2/new_images/downloaded_images'
text_folder = 'D:/Frank/8-2/new_images/downloaded_images'  # Assuming text files are in this folder
output_folder = 'D:/Frank/Input'

# Call the function
create_pickle_files(image_folder, text_folder, output_folder)
```

```python

#convert jpgs to .p files
import os
import pickle
import numpy as np
from PIL import Image

def create_pickle_files(image_folder, text_folder, output_folder):
    # Ensure the output folder exists
    os.makedirs(output_folder, exist_ok=True)

    # Loop through all files in the image folder
    for filename in os.listdir(image_folder):
        if filename.endswith(('.png', '.jpg', '.jpeg')):  # Add other extensions if needed
            # Load the image
            image_path = os.path.join(image_folder, filename)
            image = np.array(Image.open(image_path))

            # Extract the base filename without extension
            base_filename = os.path.splitext(filename)[0]

            # Read the corresponding text file
            text_file_path = os.path.join(text_folder, f'{base_filename}.txt')
            if os.path.exists(text_file_path):
                with open(text_file_path, 'r') as text_file:
                    text = text_file.read().strip()
            else:
                print(f'Text file not found for image {filename}. Skipping...')
                continue

            # Create the data dictionary
            data = {'img': image, 'text': text, 'key': base_filename}

            # Save to .p file
            pickle_path = os.path.join(output_folder, f'{base_filename}.p')
            with open(pickle_path, 'wb') as file:
                pickle.dump(data, file)
            
            print(f'Created: {pickle_path}')

# Define the image folder, text folder, and output folder
image_folder = 'D:/Frank/8-2/new_images/.pfile'
text_folder = 'D:/Frank/8-2/new_images/downloaded_images'  # Assuming text files are in this folder
output_folder = 'D:/Frank/Input'

# Call the function
create_pickle_files(image_folder, text_folder, output_folder)
```

```python
import os
import pickle

def open_top_five_p_files(directory):
    # List all files in the directory
    files = os.listdir(directory)
    
    # Filter out files that don't end with .p
    p_files = [file for file in files if file.endswith('.p')]
    
    # Sort the .p files by name
    p_files.sort()
    
    # Take the top five .p files
    top_five_p_files = p_files[:20]
    
    # List to hold the key and text content
    opened_files_content = []
    
    # Open each of the top five files and store their key and text content in the list
    for file in top_five_p_files:
        file_path = os.path.join(directory, file)
        with open(file_path, 'rb') as f:
            content = pickle.load(f)
            # Extract the key and text
            if 'key' in content and 'text' in content:
                key_text = {'key': content['key'], 'text': content['text']}
                opened_files_content.append(key_text)
    
    return opened_files_content

# Define the directory path
directory_path = 'D:/Frank/8-2/new_images/output'

# Call the function and store the result
content = open_top_five_p_files(directory_path)

# Print the result
for item in content:
    print(f"Key: {item['key']}, Text: {item['text']}")
```

```python
#convert the generated .p files back to jpg
import os
import pickle
import numpy as np
from PIL import Image
import torch

def process_and_save_images(input_directory, output_directory):
    # Ensure the output directory exists
    os.makedirs(output_directory, exist_ok=True)

    # Get a list of all pickle files in the input directory
    pickle_files = [f for f in os.listdir(input_directory) if f.endswith('.p')]

    # Process each pickle file
    for pickle_file in pickle_files:
        # Load the data from the pickle file
        with open(os.path.join(input_directory, pickle_file), 'rb') as file:
            data = pickle.load(file)

        # Check if the 'key' is in the data dictionary
        if 'key' not in data:
            print(f"'key' not found in {pickle_file}. Skipping...")
            continue

        # Check the type of data['img']
        if isinstance(data['img'], torch.Tensor):
            data['img'] = data['img'].cpu().numpy()

        if not isinstance(data['img'], np.ndarray):
            data['img'] = np.array(data['img'])

        # Ensure the data is in the correct format
        img_array = data['img']
        if img_array.ndim == 4 and img_array.shape[0] == 1:
            img_array = img_array[0]  # Remove the batch dimension
        if img_array.ndim == 3 and img_array.shape[0] == 1:
            img_array = img_array[0]  # Remove the channel dimension if grayscale
        elif img_array.ndim == 3 and img_array.shape[0] in [3, 4]:
            img_array = np.transpose(img_array, (1, 2, 0))  # Reorder dimensions to (H, W, C)

        # Convert to uint8 if necessary
        if img_array.dtype != np.uint8:
            img_array = (img_array * 255).astype(np.uint8)

        # Create an image from the numpy array
        image = Image.fromarray(img_array)

        # Extract the text and key
        text = data['text']
        key = data['key']

        # Save the image with the original filename
        output_path = os.path.join(output_directory, f'{key}.png')
        image.save(output_path)

        # Print the text
        print(f"Text for image '{key}': {text}")
        print(f"Image saved to {output_path}")

# Define input and output directories
input_directory = 'D:/Frank/8-2/new_images/output'
output_directory = 'D:/Frank/8-2/new_images/outputjpg'

# Process and save images
process_and_save_images(input_directory, output_directory)

```

```python
#extract the captions from the clip searched .p files and store them with keys
import os
import pickle

def extract_text(input_directory, output_directory):
    # Ensure the output directory exists
    os.makedirs(output_directory, exist_ok=True)

    # Get a list of all pickle files in the input directory
    pickle_files = [f for f in os.listdir(input_directory) if f.endswith('.p')]

    # Process each pickle file
    for pickle_file in pickle_files:
        # Load the data from the pickle file
        file_path = os.path.join(input_directory, pickle_file)
        with open(file_path, 'rb') as file:
            data = pickle.load(file)

        # Check if the 'key' and 'text' are in the data dictionary
        if 'key' in data and 'text' in data:
            key = data['key']
            text = data['text']
            txt_filename = f"{key}.txt"
            txt_file_path = os.path.join(output_directory, txt_filename)
            
            # Save the text information to a .txt file
            with open(txt_file_path, 'w') as txt_file:
                txt_file.write(text)
            
            print(f"Created: {txt_file_path}")
        else:
            print(f"Missing 'key' or 'text' in file: {file_path}")

# Define the input and output directories
input_directory = 'D:/Frank/8-2/new_images/output'
output_directory = 'D:/Frank/8-2/new_images/outputjpg'

# Call the function to save text information as .txt files
extract_text(input_directory, output_directory)
```
